```python
import csv
import numpy as np
import copy
```

```python
#open the cvs
with open('game_list.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    
print(data)
```

```python
#make list of teams
teams = ['Adelaide', 'Brisbane Lions', 'Carlton', 'Collingwood', 'Essendon',
 'Fremantle', 'Geelong', 'Gold Coast', 'Greater Western Sydney', 'Hawthorn', 'Melbourne',
'North Melbourne', 'Port Adelaide', 'Richmond', 'St Kilda', 'Sydney', 'West Coast', 'Western Bulldogs' ]
```

```python
#find all games with matching team name
team_list = []
for team in teams:
    for game in data:
        if (game[0] == team or game[2] == team):
            team_list.append(game)
            
    with open('team_lists/'+team+'.csv', 'w', newline='') as myfile:
        wr = csv.writer(myfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_ALL)
        for row in team_list:
            wr.writerow(row)
    
    team_list.clear()
```
